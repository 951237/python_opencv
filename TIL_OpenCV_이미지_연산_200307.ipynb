## openCV를 활용한 이미지 크기 및 픽셀 정보 확인

## openCV를 활용한 특정 범위 픽셀 변경


```
import cv2
```

```
import matplotlib.pyplot as plt
import time
```

```
image = cv2.imread('cat.png')
start_time = time.time() # 시간측정 시작
for i in range(0, 100):
    for j in range(0, 100):
        image[i, j] = [255, 255, 255] # 흰색으로 바꾸기
print(f'--- {time.time() - start_time} second ---')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) # 이미지를 BGR에서 RGB로 변환하기 
plt.show()
```

```
image = cv2.imread('cat.png')
start_time = time.time()
image[0:100, 0:100] = [0, 0, 0]
print(f'--- {time.time() - start_time} second ---')

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

## OpenCV를 활용한 ROI(Region of Interest : 관심 있는 영역) 추출

```
import cv2
import matplotlib.pyplot as plt
```

```
image = cv2.imread('cat.png')

#Numpy Slicing : ROI 처리 기능
roi = image[200:550, 50:400] # 200:550 y 축에서 200부터 550까지, 50:400 X축에서 50부터 400까지의 범위

#ROI 단위로 이미지 복사하기
image[100:450, 0:350] = roi # y축의 범위, x축의 범위에 붙여넣기

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

## OpenCV를 활용한 픽셀별 색상 다루기

```
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('cat.png')
image[:, :, 2] = 0 # 0 = 노랑, 1 = 빨강, 2 = 초록
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
